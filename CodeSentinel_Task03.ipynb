```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# For setting styles
sns.set(style = "whitegrid")

# Load the dataset
df = pd.read_csv('C:/Users/PMLS/OneDrive/Documents/netflix_titles/netflix_titles.csv')
print(df.head())
```

### 1. Correlation Matrix Heatmap

```python
# Convert 'date_added' to datetime and extract year/month
df['date_added'] = pd.to_datetime(df['date_added'].astype(str).str.strip(), errors = 'coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

# Now check correlation amomg numerical columns
correlation = df[['release_year', 'year_added', 'month_added']].corr()

# Plot correlation matrix heatmap
plt.figure(figsize = (8, 6))
sns.heatmap(correlation, annot = True, cmap = 'coolwarm', linewidths = 0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### 2. Top 5 most frequent values in selected columns

```python
print("Top 5 Countries:")
print(df['country'].value_counts().head(5))

print("\nTop 5 Genres:")
print(df['listed_in'].value_counts().head(5))

print("\nTop 5 Directors:")
print(df['director'].value_counts().head(5))
```

### 3. Distribution plots (histograms, boxplots)

```python
# Histogram of 'release_year'
plt.figure(figsize = (10, 6))
sns.histplot(df['release_year'], bins = 30, kde = True, color = 'skyblue')
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot of shows/movies added by year
plt.figure(figsize = (10, 6))
sns.boxplot(x = 'type', y = 'year_added', data = df)
plt.title('Year Added by Type (Movies vs TV Show)')
plt.show()
```
